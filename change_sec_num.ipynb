# 错开上面的corner case，来做编号的变化

```python
# edit sec num in tables!
counter = 0
for table_index in tqdm(range(len(list(doc.tables)))[:]):
    table = doc.tables[table_index]
    for row in table.rows:
        for cell in row.cells:
            for para in cell.paragraphs:
                
                it = utils.get_sec_num_iter(para.text)
                for i in it:
                    sec_num = i.group(0)
                    if sec_num == skip:
                        break
                    tmp = [int(x) for x in sec_num.split('.')]
                    head_num = tmp[0]
                    if head_num >= 4 and head_num <= 28:
                        tmp[0] = head_num + 1 if INCRE else head_num - 1
                        new_sec_num = '.'.join([str(x) for x in tmp])
                        # do replace
                        utils.replace_para_text(para,sec_num,new_sec_num)
                        counter += 1
print(counter)
```
