```python
# 测试下输出
for para in doc.paragraphs:
    txt = para.text
    sec_num = utils.get_sec_num_new(txt, from_head=True)
    
    if sec_num is None:
        continue
    
    print(txt)
```

```python
# 第一位序号是否是自增的
last = 0
for para in doc.paragraphs:
    txt = para.text
    sec_num = utils.get_sec_num_new(txt, from_head=True)
    
    if sec_num is None:
        continue
    now = int(sec_num.split('.')[0])
    if now < last:
        print(txt)
    last = now
    
# 如无输出，是严格自增的
```

```python
# 找小节title，输出所在level和当前的style（注意有的是非法）
map_level2style = {}
for para in doc.paragraphs:
    txt = para.text
    sec_num = utils.get_sec_num_new(txt, from_head=True)
    
    if sec_num is None:
        continue
    
    style = utils.get_style_name(para.style)
    level = len(sec_num.split('.')) # 几级标题
    
    if level not in map_level2style.keys():
        map_level2style[level] = {}
    if style not in map_level2style[level].keys():
        map_level2style[level][style] = 0
    map_level2style[level][style] += 1
```
